```python
""" 3 对字符串进行翻转
题目描述: 实现字符串的反转， 要求不适用任何系统方法，且时间复杂度最小
分析解答：通过字符的交换实现，需要把字符串转换为字符数组，然后定义两个索引分别指向数组的首尾，
            再交换两个索引位置的值，同时把两个索引的值想中间移动，直到两个索引相遇为止，则完成了字符串的反转。
            根据字符交换方法的不同，采用以下两种方法
方法一：临时变量法   定义一个中间变量来交换两个值。
思路：例如要交换a与b，通过定义一个中间变量temp来实现变量的交换
    temp = a;
    a = b;
    b = temp;

方法二：直接交换法   在交换变量时采用异或的特性来实现。
该方法基于如下特性：a^a = 0, a^0 = a 以及异或操作满足交换律与结合律，若要交换两个变量a和b，则可采用如下方法实现：
    a = a^b
    b = a^b     // b = a^b = (a^b)^b = a^(b^b) = a^0 = a
    a = a^b     // a = a^b = (a^b)^a = (b^a)^a = b^(a^a) = b^0 = b

"""

# 方法1 实现
def reverseStr(str):
    ch = list(str)
    lens = len(ch)
    i = 0
    j = lens -1

    while i < j:
        tmp = ch[i]
        ch[i] = ch[j]
        ch[j] = tmp
        i += 1
        j -= 1
    return ''.join(ch)


# 方法2 直接交换方法实现
def reverseStrTwo(str):
    ch = list(str)
    lens = len(ch)
    i = 0
    j = lens -1
    while i < j:
        # python 中不能直接对字符串进行异或操作，所以借助ord和chr函数
        ch[i] = chr(ord(ch[i])^ord(ch[j]))
        ch[j] = chr(ord(ch[i])^ord(ch[j]))
        ch[i] = chr(ord(ch[i])^ord(ch[j]))

        i += 1
        j -= 1
    return ''.join(ch)



if __name__ == "__main__":
    str = "abcdefg"
    print("字符串"+str+"使用方法1反转后为：",end='')
    print(reverseStr(str))
    print("字符串"+str+"使用方法2反转后为：",end='')
    print(reverseStrTwo(str))
```

```python
"""4.判断两个字符串是否为换位字符串
题目描述：换位字符串是指字符串的字符相同，但位置不同。例如：字符串"aaaabbc"与"abcbaaa"就是换位字符串。
分析解答：空间换时间，额外增加存储空间来达到优化性能目的。
        假定只是用ASCII字符，由于ASCII字符共有256个（对应编码0~255），在实现时可通过申请大小为256的数组来记录各个字符出现的个数，并将其初始化为0，然后遍历第一个字符串，将字符串对应的ASCII码值作为数组下标，把对应数组的元素加1，然后遍历第二个字符串，把数组中对应的元素值减1.
        若最后数组中各个元素的值都为0，那么这两个字符串是由相同的字符组成；否则，两个字符串是由不同字符所组成。
"""
def compareString(str1,str2):
    """
    方法功能：判断两个字符串是否为换位字符串
    输入参数：str1，str2为两个字符串
    返回值：如果时返回true，否则返回false
    """
    result = True
    bCount = [None] * 256
    i = 0
    while i < 256:
        bCount[i] = 0
        i += 1
    i = 0
    while  i <len(str1):
        bCount[ord(list(str1)[i]) - ord('0')] += 1
        i += 1
    i = 0
    while i <len(str2):
        bCount[ord(list(str2)[i]) - ord('0')] -= 1
        i += 1
    i = 0
    while i <256:
        if bCount[i] != 0:
            result = False
            break
        i += 1
    return result

if __name__ == "__main__":
    str1 = "aaabbc"
    str2 = "abcbaa"
    print("字符串1 和 字符串2分别为: %s , %s" %(str1,str2))
    if compareString(str1,str2):
        print("是换位字符.")
    else:
        print("不是换位字符。")
```

```python
"""5.判断两个字符串的包含关系
题目描述：由字母组成的字符串s1，s2，其中s2中字母个数少于s1。当出现在s2中字符在s1中都存在。
        例如 s1=”abcdef"， s2="acf"， 那么s1就包含s2；如果s1="abcdef"， s2="acg"，
        那么s1就不包含s2，因为s2中有'g', 但是s1中没有'g'
分析与解答：
方法1：直接法（时间换空间）
    对s2中的每个字符，遍历字符串s1查看是否包含字符。

方法2：（空间换时间）
    首先，定义一个flag数组来记录较短的字符串中字符出现的情况，如果出现，那么标记为1，否则标记为0，
    同时记录flag数组中1的个数count；接着遍历较长的字符串，对于字符a，
    若原来flag[a] == 1，则修改flag[a] = 0,并将count减少1；
    若flag[a] == 0, 则不做处理。最后判断count的值，
    如果count==0， 那么说明这两个字符有包含关系。
"""
# 方法 1
def isContain(str1,str2):
    len1 = len(str1)
    len2 = len(str2)

    # 字符串1比字符串2短
    if len1 < len2:
        i = 0
        while i <len1:
            j = 0
            while j <len2:
                if (list(str1)[i]  == list(str2)[j]):
                    break
                j += 1
            if j >= len2:
                return False
            i += 1
    else:
        # 字符串1比字符串2长
        i = 0
        while i<len2:
            j = 0
            while j < len1:
                if (list(str1)[j] == list(str2)[i]):
                    break
                j += 1
            if j >= len1:
                return False
            i += 1
    return True


# 方法 2
def isContainTwo(str1,str2):
    k = 0   # 字母对应数组的下标
    # 用来记录52个字母出现的情况
    flag = [None] * 52
    i = 0
    while i <52:
        flag[i] = 0
        i += 1
    count = 0       # 记录短字符串中不同字符出现的个数
    len1 = len(str1)
    len2 = len(str2)

    # shorStr, longStr 分别用来记录较短和较长的字符串
    # maxLen, minLen 分别用来记录较长和较短字符串的长度
    if len1 < len2:
        shortStr = str1
        longStr = str2
        minLen = len1
        maxLen = len2

    else:
        shortStr = str2
        longStr = str1
        minLen = len2
        maxLen = len1
    
    # 遍历短字符串
    i = 0
    while i< minLen:
        # 把字符串转换成数组对应的下标（大写字母0~25， 小写字母26~51）
        if ord(list(shortStr)[i]) >= ord('A') and ord(list(shortStr)[i]) <= ord('Z'):
            k = ord(list(shortStr)[i]) - ord('A')
        else:
            k = ord(list(shortStr)[i]) - ord('a') + 26
        if flag[k] == 0:
            flag[k] =1 
            count += 1
        i += 1
    
    # 遍历长字符串
    j = 0
    while j < maxLen:
        if ord(list(longStr)[j]) >= ord('A') and ord(list(longStr)[j]) <= ord('Z'):
            k = ord(list(longStr)[j]) - ord('A')
        else:
            k = ord(list(longStr)[j]) - ord('a') + 26
        if flag[k] == 1:
            flag[k] = 0
            count -= 1
            if count == 0:
                return True
        j += 1
    return False



if __name__ == '__main__':
    str1 = 'abcderf'
    str2 = 'acf'

    print("使用isContain 函数")
    isContainValue = isContain(str1,str2)
    print("字符串： %s 和 字符串：%s"%(str1,str2))
    if (isContainValue):
        print("有包含关系。")
    else:
        print("没有包含关系。")
    

    print("使用isContainTwo函数")
    isContainValue = isContainTwo(str1,str2)
    print("字符串： %s 和 字符串：%s"%(str1,str2))
    if (isContainValue):
        print("有包含关系。")
    else:
        print("没有包含关系。")
```

```python
"""6.对由大小写字母组成的字符数组排序
题目描述：一个由大小写字母组成的字符串，对他进行重新组合，使得其中所有小写字母排在大写字母前
        (大写字母或小写字母之间不要求保持原有次序)
分析解答：使用类似快排方法处理。用两个索引分别指向字符串的首尾，首索引正向遍历字符串，
        找到第一个大写字母，尾索引逆向遍历字符串，找到第一个小写字母交换两个索引位置的字符，
        然后将两个索引沿着相应的方向继续向前移动，重复，直到受索引大于或等于尾索引为止。

"""
def ReverseArray(ch):
    """
    函数功能：对字符数组排序，使得小写字母在前，大写字母在后
    """
    lens = len(ch)
    begin = 0
    end = lens -1
    while begin < end:
        # 正向遍历找到下一个大写字母
        while ch[begin] >= 'a' and ch[end] <= 'z' and end > begin:
            begin +=1 
        # 逆向遍历找到下一个小写字母
        while ch[end] >= 'A' and ch[end] <= 'Z' and end >begin:
            end -= 1
        temp = ch[begin]
        ch[begin] = ch[end]
        ch[end] = temp

if __name__ == '__main__':
    ch = list("ABcDefGhi")
    ReverseArray(ch)
    i = 0
    while i < len(ch):
        print("--", ch[i],end='')
        i += 1
```

```python
"""7.消除字符串的内嵌括号
题目描述：给定如下格式的字符串: (1,(2,3),(4,(5,6),7)), 括号内的元素可以是数字，也可以是另一个括号，实现一个算法消除嵌套的括号。
            例如把 (1,(2,3),(4,(5,6),7))变为(1,2,3,4,5,6,7)
分析解答：这里主要实现两个功能：1.判断表达式是否正确；2.消除表达式中的嵌套括号。针对第一个功能考虑从一下方面入手：首先，判断表达式中数字、逗号、括号字符，如果出现其他字符，就是非法表达式；
其次，判断括号是否匹配，如果碰到'('，那么把括号的计数器的值+1；如果')',，那么计数器值减一，否则位非法表达式，当遍历表达式完毕后，括号计数器的值位0， 说明括号是成对出现的，否则，括号不配对，表达式为非法表达式。
针对第二个功能：可通过申请一个额外的存储空间，在遍历原字符串的时候把除括号以外字符保存到新申请的额外存储空间中，即可去掉嵌套括号。
要注意：字符串首尾括号保存好。
"""

def removeNestedParas(strs):
    """ 
    方法功能：去掉字符串中嵌套的括号
    """
    if strs == None:
        return strs
    parenthes_num  = 0  # 记录不匹配'(' 出现的次数
    
    if list(strs)[0] != '(' or list(strs)[-1] != ')':
        return None

    sb = '('
    i = 1
    while i <len(strs) -1:
        ch = list(strs)[i]
        if ch == '(':
            parenthes_num += 1
        elif ch == ')':
            parenthes_num -= 1
        else:
            sb = sb + (list(strs)[i])

        i += 1
    
    # 判断括号是否匹配
    if parenthes_num != 0:
        print("括号不匹配。")
        return None
    sb = sb +')'
    return sb

if __name__ == "__main__":
    strs = "(1,(2,3),(4),(5,6,7,8,(10)))"
    print(strs + "去除括号后为：" + removeNestedParas(strs))
```

```python
"""9.实现字符串的匹配
题目描述：给主字符串S与模式字符串P， 判断P是否是S的子串，如果是，那么找出P在S中第一次出现的下标
分析解答：对于字符串的匹配，直接方法是逐个比较字符串中的字符，该方法比较易实现但效率低下。对于字符串匹配问题，经典的KMP算法效率较高。
"""
"""方法1：直接计算法
假定主串S=s0,s1,s2..sm, 模式串P=p0,p1,p2,...,pn.实现方法 比较主串S中以S_i 为首的字符串和模式串P，判断P是否为S的前缀，若是，那么P在S中第一次出现的位置则是i，否则接着比较从
S_i+1 开始的子串与模式串P，该方法时间复杂度为O(m+n). 此外如果 i>m-n, 那么主串中以S_i 为首的子串长度必定小于模式串P的长度。
"""
def matchStringP(s,p):
    """
    方法功能：判断p是否为s的子串，如果是，那么返回P在s中第一次出现的下标，否则返回-1
    输入参数：s和p分别为主串和子串
    """
    #检验参数的合理性
    if s == None or p == None:
        print("参数不合理")
        return -1
    slen = len(s)
    plen = len(p)

    # p 肯定不是s的子串
    if slen < plen:
        return -1
    i,j = 0,0
    while i <slen and j < plen:
        if list(s)[i] == list(p)[j]:
            # 如果相同，那么继续比较后面的字符
            i += 1
            j += 1
        else:
            # 后退回去重新比较
            i = i -j +1
            j = 0
            if (i > slen - plen):
                return -1
    if j >= plen: # 匹配成功
        return i - plen
    return -1
"""该方法在最差情况下需要对模式串P遍历m-n次，（m，n分别为主串和模式串的长度），因此算法复杂度为O(n(m-n))"""

"""方法2KMP算法 
在 p0,p1,p2...pj-1  == Si-j ... Si-1 中，模式串的前j个字符已经和主串中的i-j到i-1的字符串进行了比较，此时如果Pj != Si, 那么模式串需要回退到0，主串需要回退到i-j + 1的位置重新开始下一次比较。
在KMP算法中， 如果Pj != Si 那么不需要回退，即i保持不动， j也不用清零，二十享有华东模式串，用Pk 和Si继续匹配。方法的核心是确定k的大小。
"""
def getNext(p,nexts):
    """
    方法功能：求字符串的next数组
    输入参数：p为字符串，nexts为P的next数组
    """
    i = 0
    j = -1
    nexts[0] = -1
    while i < len(p):
        if j == -1 or list(p)[i] == list(p)[j]:
            i += 1
            j += 1
            nexts[i] = j
        else:
            j = nexts[j]

def matchStringP2(s,p,nexts):
    # 检验参数的合理性，s的长度一定不小于p的长度
    if s == None or p == None:
        print("参数不合理")
        return -1
    slen = len(s)
    plen = len(p)

    # p肯定不是s的子串
    if slen < plen:
        return -1

    i = 0
    j = 0
    while i < slen and j < plen:
        print("i=" + str(i) + ", j =" +str(j))
        if j == -1 or list(s)[i] == list(p)[j]:
            # 如果相同，那么继续比较后面的字符
            i += 1
            j += 1
        else:
            # 主串i不需要回溯，从next数组中找出需要比较的模式串的位置j
            j = nexts[j]
    if j >= plen: # 匹配成功
        return i-plen
    return -1


if __name__ == "__main__":
    s = "xyzabcdef"
    p = "abc"
    # print(matchStringP(s,p))
    
    s2 = "abababababcbabddd"
    p2 = "abaabc"
    lens = len(p2)
    nexts = [0]*(lens + 1)
    getNext(p2, nexts)

    print("next 数组为:" + str(nexts[0]))
    i= 1
    while i <lens -1:
        print(","+str(nexts[i]))
        i += 1
    print("\n")
    print("匹配结果为：" + str(matchStringP2(s2,p2,nexts)))

```

```python
"""10.求字符串里的最长回文子串
题目描述：回文子串是指一个字符串从左到右与从右往左便利的到的序列相同。
分析思路：最容易想到的思路是遍历字符串的所有可能子串，判断是否为回文字符串，然后找出最长回文子串。当字符串很长时，该方法效率低下，方法不可取。
"""
"""方法1： 动态规划方法
蛮力法查找时候徐多字符的比较是重复的，因此可把前面比较的中间结果记录下来以后使用。
给定字符串S0,S1,S2,...Sn 。假设P(i,j) = 1 表示Si, Si+1, ..., Sj 是回文字符串； P(i,j) = 0 ,则P(i,j) = 1;
如果Si == Si+1: 那么P(i,i+1) = 1, 否则P(i,i+1) = 0 
如果Si == Sj：   那么P(i,j) = P(i+1,j-1) 
"""
class Test:
    def __new__(self):
        self.startIndex = None
        self.lens = None
    def getStartIndex(self):return self.startIndex
    def getLen(self):return self.lens

    def getLongestPalindrome(self,strs):
        """
        方法功能：找出字符串中最长的回文子串
        输入参数：str为字符串，startIndex与len为找到的回文字符串的起始位置与长度
        """
        if strs == None:
            return 
        n = len(strs)   # 字符串长度
        if n < 1:
            return 
        self.startIndex = 0
        self.lens = 1
        
        # 申请额外的存储空间记录查找的历史信息
        historyRecord = [([None] * n) for i in range(n)]
        i = 0
        while i <n:
            j = 0
            while j <n:
                historyRecord[i][j] = 0
                j += 1
            i += 1
        # 初始化长度为1的回文字符串信息
        i = 0
        while i < n:
            historyRecord[i][j] = 1
            i += 1
        # 初始化长度为2 的回文子串信息
        i = 0
        while i < n-1:
            if list(strs)[i] == list(strs) [i+1]:
                historyRecord[i][i+1] = 1
                self.startIndex = i
                self.lens = 2
            i += 1
        # 查找从长度为3开始的回文字符串
        pLen = 3
        while pLen <= n:
            i = 0
            while i <n - pLen + 1:
                j = i + pLen -1
                if list(strs)[i] == list(strs)[j] and historyRecord[i + 1][j-1] == 1:
                    historyRecord[i][j] = 1
                    self.startIndex = i
                    self.lens = pLen
                i += 1
            pLen += 1


"""方法2： 中心扩展法
思路：判断一个字符串是否为回文字符串最简单的方法是：从字符串最中间的字符开始向两边扩展，通过比较左右两边字符侍奉相等就可以确定这个字符串是否为回文字符串。
这种方法对于长度为奇数和偶数的字符串需要分别对待。

"""
class Test2:
    def __init__(self):
        self.startIndex = None
        self.lens = None
    
    def getStartIndex(self):
        return self.startIndex

    def getLen(self):
        return self.lens
    
    def expandBothSide(self, strs, c1,c2):
        """
        方法功能：对字符串str，以c1 和c2为中心向两侧扩展寻找回文子串
        """
        n = len(strs)
        while c1 >= 0 and c2 <n and list(strs)[c1] == list(strs)[c2]:
            c1 -= 1
            c2 += 1
            tmpStartIndex = c1 + 1
            tmpLen = c2 - c1 - 1

            if tmpLen > self.lens:
                self.lens = tmpLen
                self.startIndex = tmpStartIndex

    def getLongestPalindrome(self, strs):
        """
        方法功能：找出字符串最长的回文子串
        """
        if strs == None:
            return
        n = len(strs)
        if n < 1:
            return
        i = 0

        while i <n-1:
            # 找回文子串长度为奇数时， 从第i个字符向两边扩展
            self.expandBothSide(strs, i,i)
            # 偶数，从第i和i+1两个字符向两边扩展
            self.expandBothSide(strs,i,i+1)
            i += 1


if __name__ == "__main__":
    # strs = "abcdefgfedxyz"
    # t = Test()
    # # t.getLongestPalindrome(strs)
    # t.getLongestPalindrome(strs)
    # if t.getStartIndex() != -1 and t.getLen() != -1:
    #     print("最长的回文字符串为：")
    #     i = t.getStrtIndex()
    #     while i <t.getStartIndex() + t.getLen():
    #         print(list(strs)[i])
    #         i += 1
    # else:
    #     print("查找失败")
    str2 = "abcedfgfedxyz"
    t2 = Test2()
    t2.getLongestPalindrome(str2)
    if t2.getStartIndex() != -1 and t2.getLen() != -1:
        print("最长回文子串为：")
        i = t2.getStartIndex()
        while i <t2.getStartIndex() + t2.getLen():
            print(list(str2)[i])
            i += 1
    else:
        print("查找失败")
```

```python
"""12.判断字符串中是否包含重复字符
题目描述：判断一个字符串中是否包含重复字符。例如'google' 就包含重复字符o, 'abc' 中就不包含重复字符
"""

"""方法1： 蛮力法
把字符串看成一个字符数组， 对该数组使用双重循环进行遍历，即对每个字符都将其于后面所有字符进行比较，如果找到相同字符，则说明重复
"""
def isDup(strs):
    # 函数功能：判断字符串中是否有重复字符
    lens = len(strs)
    i = 0
    while i < lens:
        j = i + 1
        while j < lens:
            if list(strs)[j] == list(strs)[i]:
                return True
            j += 1
        i += 1
    return False


"""方法2：空间换时间
主要思路：常见字符只有256个， 假设该题涉及的字符串中不同的字符个数最多为256个， 
那么申请大小为256的int类型数组来记录每个字符出现的次数，初始化为0， 把这个字符的编码作为数组的下标，在遍历字符数组时，若该字符在数组中对应值0，那么设置为1
如果为1，那么说明这个字符在前面已经出现，因此，字符串包含重复的字符。采用该方法只需要对字符数组进行一次遍历即可，因此时间复杂度为O(N), 但需要额外申请256个单位的空间。
由于申请的数组用来记录一个字符是否出现，只需要1bit也能实现该功能，因此作为更好方案，可只申请大小为8的int类型的数组， 由于每个int类型占32bit，所以，大小为8的数组总共为256bit
用1bit来表示一个字符是否已经出现过可达到同样目的。
"""
def isDup2(strs):
    lens = len(strs)
    flags = [None] * 8 # 只需要8个 32 位的int， 8 * 32 = 256位
    i = 0
    while i < 8:
        flags[i] = 0
        i += 1
    i = 0
    while i < lens:
        index = ord(list(strs)[i]) / 32
        shift = ord(list(strs)[i]) % 32
        if (flags[index] & (1<<shift)) != 0:
            return True
        flags[index] |= ( 1<< shift)
    return False



if __name__ == "__main__":
    # strs = "google"
    # result = isDup(strs)
    # if result == True:
    #     print(" is Dup")
    # else:
    #     print("is not Dup")

    strs2 = "google"
    results2 = isDup2(strs2)
    if results2 :
        print(strs2 + "中有重复字符")
    else:
        print( strs2 + "is not Dup") 
```

```python
"""13.找到由其他单词组成的最长单词
题目描述：给定一个字符串数组，找出数组中最长的字符串，使其能有数组中其他的字符串组成。"""

class LongestWord:
    def find(self, strArray, strs):
        """方法功能：判断字符串strs是否在字符串数组中"""
        i = 0
        while i < len(strArray):
            if strs == strArray[i]:
                return True
            i += 1
        return False

    def isContain(self,strArray, word, length):
        """方法功能：判断字符串word是否能由数组strArray中的其他单词组成
        函数参数：word为待判断的后缀子串；length为待判断字符串的长度。
        """
        lens = len(word)

        # 递归结束的条件是：当字符串长度为0时，说明该字符串已经遍历完成
        if lens == 0:
            return True
        
        # 循环取字符串的所有前缀
        i = 1
        while i <= lens:
            # 取到的子串为自己
            if i == length:
                return False
            strs = word[0:i]
            if self.find(strArray, strs):
                # 查找玩字符串的前缀后， 递归判断后面的子串是否能有其他单词组成
                if self.isContain(strArray, word[i:], length):
                    return True
            i += 1

        return False

    def getLongestStr(self, strArray):
        """方法功能：找出能由数组中其他字符串组成的最长字符串"""
        # 对字符串由大到小排序
        strArray = sorted(strArray, reverse=True)
        print(strArray)

        # 贪心地从最长的字符串开始判断
        i = 0
        while i < len(strArray):
            if self.isContain(strArray, strArray[i], len(strArray[i])):
                return strArray[i]
            i += 1
        # 如果没有找到，那么就返回空串
        return None

if __name__ == "__main__":
    strArray = ["test", "tester","testertest","testing","apple", "seattle","banana", "batting", "ngcat", "batti", "bat", "testingtester" "testbattingcat"]
    lw = LongestWord()
    logestStr = lw.getLongestStr(strArray)
    if logestStr != None:
        print("the longest string is :" + logestStr)
    else:
        print("not exist this string")

"""排序的时间复杂度为O(nlogn), 假设单词的长度为m，那么有m中前缀，判断一个单词是否在数组中的时间复杂度为O(mn), 由于总共有n个字符串， 因此判断所需的时间复杂度为O(m*n^2). 因此总的时间复杂度为O(nlogn + m*n^2)."""
```

```python
"""14.统计字符串中连续的重复字符个数
题目描述：使用递归方法实现一个 字符串中连续出现相同字符的最大值，例如字符串“aaabbcc”， 中连续出现字符‘a’的最大值是3，字符串‘abbc’中连续出现字符b的最大值是2
"""
def getMaxDupChar(s,startIndex, curMaxLen, maxLen):
    # 字符串遍历结束，返回最长连续重复字符串的长度
    if startIndex == len(s) -1:
        return max(curMaxLen, maxLen)
    # 如果两个连续的字符相等，那么在递归调用的时候把当前最长串的长度+1
    if list(s)[startIndex] == list(s)[startIndex + 1]:
        return getMaxDupChar(s, startIndex+1, curMaxLen+1, maxLen)
    
    # 两个连续子串不相等，求出最长串max(curMaxLen, maxLen)
    # 当前连续重复字符串的长度变为1
    else:
        return getMaxDupChar(s, startIndex + 1, 1, max(curMaxLen, maxLen))
    
if __name__ == "__main__":
    print ("abbc max contain string length is: " + str(getMaxDupChar("abbc", 0,1,1)))
```

```python
"""15. 求最长递增子序列的长度
题目描述:假设L = <a1,a2,...an>是n个不同的实数的序列，L的递增子序列是这样一个子序列Lin = <ak1,ak2,...,akm>，其中，k1 < k2 < ... < km 且 ak1 < ak2 < ... < akm, 求最大值m
"""

"""方法1： 最长公共子串法
对序列L = <a1, a2, ..., an> 按递增进行排序得到序列Lo = <b1, b2, ..., bn> 。显然， L与LO的公共子序列就是L的最长递增子序列，因此可以使用求公共子序列的方法来求解。
"""

"""方法2：动态规划法
由于以地i个元素为结尾的最长递增子序列只与第i-1个元素为结尾的最长递增子序列有关。
以第i个元素为结尾的最长递增子序列的取值有两种可能：
1）第i个元素单独作为一个子串 (L[i] <= L[i-1])
2) 以第i-1 个元素为结尾的最长递增子序列加1 (L[i] > L[i-1])
由此，得到如下递推公式：假设maxLen[i] 表示以第i个元素为结尾的最长递增子序列，那么
1) maxLen[i]  = max[1, maxLen[j] + 1], j < i and L[j] < L [i]
2) maxLen[0] = 1
"""
def getMaxAscendingLen(strs):
    """函数功能：求字符串L的最长递增子串的长度"""
    lens = len(strs)
    maxLen = [None] * lens
    maxLen[0] = 1
    maxAscendingLen = 1
    i = 1
    while i < lens:
        maxLen[i]  = 1 # maxLen[i] 的最小值是1
        j = 0
        while j < i:
            if list(strs)[j] < list(strs)[i] and maxLen[j] > maxLen[i-1]:
                maxLen[i] = maxLen[j] + 1
                maxAscendingLen = maxLen[i]
            j += 1
        return maxAscendingLen

if __name__ == "__main__":
    s = "xbcdza"
    print ("max sub string lengths is: " + str(getMaxAscendingLen(s)))
```

```python
"""17.求解字符串中字典序最大的子序列
题目描述：给定一个字符串，求串中字典序最大的子序列。字典序最大的子序列是这样构造：给丁字符串a0,a1,...,an-1, 首先在字符串a0,a1,...an-1中找到值最大的字符ai， 
然后再剩余的字符串ai+1, ..., an-1中找到值最大的字符aj然后再剩余的aj+1,..., an-1中找到值最大的字符 ak。。。指导字符串的长度为0， aiajak。。即为答案。"""

def getLargestSub(src):
    """方法功能：求串中字典序最大的子序列"""
    if src == None:
        return None
    lens = len(src)
    largestSub = [None] * (lens+1)
    k,i= 0,0
    while i < lens:
        largestSub[k] = list(src)[i]
        j = i + 1
        while j < lens:
            # 找出第i个字符后面最大的字符放到largestSub[k] 中
            if list(src)[j] > largestSub[k]:
                largestSub[k] = list(src)[j]
                i = j
            k += 1
            i += 1
        return ''.join(largestSub[0:k])

if __name__ == "__main__":
    gls = "acdbxmng"
    res_gls = getLargestSub(gls)
    if res_gls == None:
        print ("string is empty")
    else:
        print(res_gls)
```

```python
"""18.判断一个字符串是否由另外一个字符串旋转得到
题目描述：给定一个能判断一个单词是否为另一个单词子字符串的方法，记为isSubstring. 如何判断s2是否能通过旋转s1得到（只能使用一次isSubstring方法），例如“waterbottle" 可以通过字符串"erbottlewat
"旋转得到。
分析思路：如果没有函数使用次数限制，可通过求出s2进行旋转的所有组合与s1进行比较。该方法时间复杂度较高。通过对字符串旋转进行分析，字符串s1旋转后得到的字符串一定是s1s1的子串。
因此通过判断s2是否是s1s1的子串来判断s2能否通过旋转s1得到。
"""
""""""
def isSubstring(str1,str2):
    """函数功能：判断str2 是否为 str1 的子串"""
    return str1.find(str2) != -1

def rotateSame(str1,str2):
    """函数功能：判断str2是否可通过旋转str1得到"""
    if str1 == None or str2 == None:
        return False
    len1 = len(str1)
    len2 = len(str2)

    # 判断两个字符串长度是否相等， 如果不相等，那么不可能通过旋转得到
    if len1 != len2:
        return False
    # 申请临时空间存储str1str1， 多申请一个空间存储'\0'
    tmp = [None] * (2 * len1 + 1)
    # tmp 为 str1str1
    i = 0 
    while i < len1:
        tmp[i] = list(str1)[i]
        tmp[i+len1]  = list(str1)[i]
        i += 1
    tmp[2 * len1] = '\0'
    # 判断str2 是否为tmp的子串
    res = isSubstring(''.join(tmp), str2)
    return res

if __name__ == '__main__':
    str1 = "waterbottle"
    str2 = "erbottlewat"
    result = rotateSame(str1, str2)
    if result:
        print(str2 + "you are right, " + str1)
    else:
        print (str2 + "you are wrong" + str1)

"""算法性能分析：
对字符串str1 进行一次遍历，时间复杂度为:O(N), 其中，N为字符串的长度， 接着调用isString函数， 假设采用KMP算法，该方法时间复杂度为O(2N+N) = O(3N), 
因此整个算法的时间复杂度为O(N).此外，这种方法申请了2N+1个存储空间，因此算法空间复杂度为O(N)"""
```

```python
"""19.求字符串的编辑距离
题目描述：编辑距离又称LEvenshtein距离， 是指两个字符串之间由一个转成另一个所需的最少编辑操作次数。许可的编辑操作包括将一个字符替换成另一个字符、插入一个字符、删除一个字符。
请设计实现一个算法来计算两个字符串的编辑距离，并计算其复杂度。再某些应用场景下，替换操作的代价比较高，假设替换操作的代价是插入和删除的两倍，算法如何调整。

分析解答：使用动态规划方法来解决，具体思路
给丁字符串s1，s2， 首先定义一个函数D(i,j) (0<= i <= strlen(s1), 0<=j<= strlen(s2)),用来表示第一个字符串s1长度为i的子串和第二个字符串s2长度为j的子串的编辑距离。从s1变成s2可通过一下三个操作完成：
1）添加操作， 假设已经计算出D(i-1,j) 的值(s1[0...i] 与s2[0...j-1]的编辑距离)， 则D(i,j) = D(i,j-1) + 1 (s1长度为i的子串后面添加s2[j] 即可)
2）删除操作， 假设已经计算出算出D(i-1,j) 的值(s1[0...i] 与s2[0...j-1]的编辑距离)，如果s1[i] = s2[j], 那么D(i,j) = D(i-1,j-1)(s1长度为i的字串删除最后的字符s1[j]即可)
3）替换操作， 假设已经计算出算出D(i-1,j-1) 的值(s1[0...i-1] 与s2[0...j-1]的编辑距离)，如果s1[i] = s2[j], 那么D(i,j) = D(i-1,j-1), 如果s1[i] != s2[j] , 那么D(i,j) = D(i-1,j-1) （替换s1[i]为s2[j]，或者s2[j] 为s1[i])


"""

"""
"""
class EditDistance:
    def mins(self,a,b,c):
        tmp = a if a < b else b
        return tmp if tmp < c else c
    
    def edit(self, s1,s2, replaceWight):
        """函数功能：参数replaceWeight用来表示替换操作与插入操作相比的倍数"""
        # 两个空串的编辑距离为0
        if s1 == None and s2 == None:
            return 0
        # 如果s1是空串，那么编辑距离应为s2的长度
        if s1 == None:
            return len(s2)
        # 如果s2是空串，编辑距离应为s1的长度
        if s2 == None:
            return len(s1)
        
        len1  = len(s1)
        len2 = len(s2)
        # 申请二位数组来存储中间的计算结果
        D = [([None] * (len2 + 1)) for i in range(len1 + 1)]
        i = 0
        while i <len1  + 1:
            D[i][0] = i
            i+= 1
        i = 0
        while i < len2 + 1:
            D[0][i] = i
            i += 1
        i = 1
        while i <len1 + 1:
            j = 1
            while j < len2 + 1:
                if list(s1)[i-1] == list(s2)[j - 1]:
                    D[i][j] = self.mins(D[i-1][j] + 1, D[i][j-1]+1, D[i-1][j-1])
                else:
                    D[i][j] = min(D[i-1][j]+1, D[i][j-1] + 1, D[i-1][j-1] + replaceWight)
                j += 1

            i += 1
        print("------------")
        i = 0
        while i < len1 + 1:
            j = 0
            while j < len2 + 1:
                print (D[i][j])
                j += 1
            print('\n')
            i += 1
        print("--------")
        dis = D[len1][len2]
        return dis

if __name__ == "__main__":
    s1 = "bciln"
    s2 = "fciling"
    ed = EditDistance()
    print("the first question:")
    print("edit distance: " + str(ed.edit(s1,s2,1)))
    print ("the second question:")
    print("edit distance:" + str(ed.edit(s1,s2,2)))

```

```python
"""20.在二维数组中寻找最短路线
题目描述：寻找一条从左上角(arr[0][0]) 到右下角(arr[m-1][n-1]) 的路线，使得沿途经过的数组中的整数的和最小。
分析思路：从右下角开始倒着分析该问题：最后一步到达arr[m-1][n-1]只有两条路：通过arr[m-2][n-1] 或 arr[m-1][n-2]到达， 假设从arr[0][0] 到 arr[m-2][n-1] 沿途径数组最小值为f(m-2,n-1), 到arr[m-1][n-2] 沿途数组最小值为f(m-1,n-2).因此最后一步选择的路线为min{f(m-2,n-1),f(m-1,n-2)}. 
同理，选择到arr[m-2][n-1] 或 arr[m-1][n-2]的路径可以采用同样的方法来确定。
"""
"""方法1：递归法
结束条件为遍历到arr[0][0].
求解过程要考虑：遍历到arr[i][j](当i=0,j=0) 时只能沿着一条固定的路径倒着往回走直到arr[0][0].
如下实现
"""
def getMinPath(arr,i,j):
    # 倒着走到了第一个结点，递归结束
    if i == 0 and j == 0:
        return arr[i][j]
    # 选择两条路可能最小值
    elif i > 0 and j > 0:
        return arr[i][j] + min(getMinPath(arr, i-1,j), getMinPath(arr, i, j-1))
    
    # 下面两个条件可以选一个
    elif i >= 0 and j == 0:
        return arr[i][j] + getMinPath(arr, i-1, j)
        # j > 0 && i == 0
    else:
        return arr[i][j] + getMinPath(arr, i, j-1)
    
def getMinPathTo(arr):
    if arr == None or len(arr) == 0:
        return 0
    return getMinPath(arr, len(arr) -1 , len(arr[0]) - 1)

if __name__ == "__main__":
    arr = [[1,4,3], [8,7,5],[2,1,5]]
    print(getMinPathTo(arr))
```
