```python
"""1.4 如何对链表进行重新排序
-题目描述：给定链表L0-L1-L2,...,Ln-1-Ln  把链表重新排序为:L0--Ln--L1--Ln-1--L2--Ln-2...    
要求：1）在原链表的基础进行排序，即不能申请新的结点；2）智能修改结点的next域，不能修改数据域


思路一：
    1）找到链表的中间结点
    2）对链表的后半部分子链表进行逆序
    3）把链表的前半部分子链表与逆序后边部分子链表进行合并，合并的思路为：分别从两个链表各取一个节点进行合并。
"""

class LNode:
    def __init__(self,x):
        self.data = x
        self.next = None

"""方法一： 按照思路一进行编码

方法功能：找出链表Head的中间结点， 把链表从中间断成两个子链表
输入参数：head:链表头结点
返回值：链表中间结点
"""
def findMiddleNode(head):
    # 1.1 判空
    if head is None or head.next is None:
        return head
    # 1.2 初始化
    fast = head         # 遍历链表时，每次走两步.快慢机的设置？
    slow = head         # 遍历链表时，每次走一步
    slowPre = head

    # 1.3 当fast走到链表尾部时，slow恰好指向链表的中间结点
    while fast is not None and fast.next is not None:
        slowPre = slow
        slow = slow.next
        fast = fast.next.next

    # 1.4 断开链表，形成两个子链表
    slowPre.next = None
    return slow


"""
方法功能：对不带头结点的单链表实现翻转
输入参数：head:链表头结点
"""
def reverse4(head):
    # 1.1 判空
    if head == None or head.next == None:
        return head
    # 1.2 初始化
    pre = head
    cur = head.next
    next = cur.next
    pre.next = None

    # 1.3 使用当前遍历到的结点cur指向其前驱节点
    while cur is not None:
        next = cur.next
        cur.next = pre
        pre = cur
        cur = cur.next

        cur = next 
    return pre

"""
f方法功能：对链表进行排序
输入参数：head链表头结点
"""
def Recorder(head):
    # 1.1 判空
    if head == None or head.next == None or head.next.next == None:
        return head

    # 1.2 前半部分链表第一个结点
    cur1 = head.next
    mid = findMiddleNode(head.next)
    # 1.3 后半部分链表逆序后第一个结点
    cur2 = reverse4(mid)
    tmp = None
    # 1.4 合并两个链表
    while cur1.next is not None:
        tmp = cur1.next
        cur1.next = cur2
        cur1 = tmp
        tmp = cur2.next
        cur2.next = cur1
        cur2 = tmp

    cur1.next = cur2

"""注意：
1.查找链表中间结点方法的时间复杂度为O(N),逆序子链表的时间复杂度也是O(N),合并两个子链表的时间复杂度也为O(N),因此，整个方法的时间复杂度为O(N)，其中N表示链表的长度，该种方法只用了常熟个额外指针变量，因此空间复杂度为O(1)
2.引申：如何查找链表中间结点？
    主要思路：用两个指针从链表的第一个结点开始同时遍历结点，一个快指针每次走两步，慢指针每次走一步，当快指针到达链表尾部，慢指针正好到达链表中部。

"""

if __name__ == "__main__":
    i = 1
    head = LNode(None)
    head.next = None
    tmp = None
    cur = head

    # 构造第一个链表
    while i < 8:
        tmp = LNode(None)
        tmp.next = None
        tmp.data  = i
        cur.next = tmp
        cur = tmp
        i += 1

    print("\n排序前:")
    cur = head.next
    while cur != None:
        print("-%d " % cur.data, end="")
        cur = cur.next

    Recorder(head)
    print("\n排序后:")
    cur = head.next
    while cur != None:
        print("-%d " % cur.data, end="")
        cur = cur.next
```

```python
"""1.5 如何找出单链表中的倒数第k个元素

题目描述：找出单链表中的倒数第k个元素。例如给定单链表：1--2--3--4--5--6--7, 则单链表中倒数第3个元素是5

方法一：顺序遍历两遍
    首先遍历一次单链表，求出单链表的整个长度n， 之后把求倒数第k个元素转换为求顺序第n-k个元素， 再去遍历一次单链表即可，
方法二：快慢指针
    从头到尾进行遍历，在查找过程中，设置两个指针，让快指针先前移k步，之后两个指针同时前移。循环直到先行指针值为None时，另一个指针所知位置就是所要找的位置
"""
class LNode:
    def __init__(self,x):
        self.data = x
        self.next = None
# 构造一个链表函数
def constructList():
    i = 1
    head = LNode(None)
    head.next = None
    tmp = None
    cur = head

    # 构造第一个链表
    while i < 10:
        tmp = LNode(None)
        tmp.data = i
        tmp.next = None
        cur.next = tmp
        cur = tmp
        i+= 1

    return head

# 顺序输出链表数据
def printList(head):
    cur = head.next
    while cur != None:
        print("-%d " % cur.data, end="")
        cur = cur.next

"""
方法功能：找出链表中倒数第k个结点
输入参数：head:链表头结点
返回值：返回倒数第k个结点
"""
def FindLastK(head, k):
    # 1.1 判空
    if head == None or head.next == None:
        return head
    # 1.2 初始化
    slow = LNode(None)
    fast = LNode(None)
    slow = head.next
    fast = head.next
    i = 0

    while i<k and fast != None:
        fast = fast.next
        i += 1

    if i < k:
        return None
    while fast != None:
        slow = slow.next
        fast = fast.next
    return slow

"""注意：
1.这里实现是采用快慢指针的方式进行实现，只需对链表进行一次遍历，因此时间复杂度为O(N),N表示链表的长度。只需要常量个指针变量来保持结点的地址信息，因此空间复杂度为O(1)
2.
"""    

if __name__ == "__main__":
    head = constructList()      # 链表头指针
    result = None
    print("链表")
    printList(head)
    result = FindLastK(head, 3)
    if result != None:
        print("\n 链表倒数第3 个元素为 %s " % (str(result.data)))
```

```python
"""1.9 如何合并两个有序链表
题目描述：已知两个链表head1和head2各自有序（例如升序），合并两个列表，要求合并后依然有序

解题思路：分别用head1和head2两个指针来遍历两个链表，如果当前head1指向的数据小于head2指向的数据，则将head1指向的结点归入合并后的链表中，否则，将head2指向的结点归入合并后的链表。
        如果一个链表遍历结束，则把未结束的链表连接到合并后链表的尾部

"""
class LNode:
    def __init__(self,x):
        self.data = x
        self.next = None
                                
# 构造链表
def constructList(start):
    i = start
    head = LNode(None)
    head.next = None
    tmp = None
    cur = head
    while i < 7:
        tmp = LNode(None)
        tmp.data = i
        tmp.next = None
        cur.next = tmp
        cur = tmp
        i += 2

    return head

# 构造输出链表函数，使用printList(head)

"""
方法功能：合并两个升序排列的单链表
输入参数：head1与head2代表两个单链表
返回值：合并后链表的头结点
"""
def Merge(head1,head2):
    # 1.1判空
    if head1 == None or head1.next == None:
        return head2
    if head2 == None or head2.next == None:
        return head1
    
    # 1.2 初始化
    cur1 = head1.next
    cur2 = head2.next
    head = None     # 合并后的头结点
    cur = None      

    # 1.3 合并后的头结点为第一个结点元素最小的那个链表的头结点
    if cur1.data < cur2.data:
        head = head1
        cur = cur1
        cur1 = cur1.next

    else:
        head = head2
        cur = cur2
        cur2 = cur2.next
    
    # 1.4 每次找链表剩余结点的最小值对应的结点连接到合并后链表的尾部
    while cur1 != None and cur2 != None:
        if cur1.data <cur2.data:
            cur.next = cur1
            cur = cur1
            cur1 = cur1.next
        else:
            cur.next = cur2
            cur = cur2
            cur2 = cur2.next
    # 1.5 当遍历完一个链表后把另一个链表剩余的结点合并到链表之后
    if cur1 != None:
        cur.next = cur1
    if cur2 != None:
        cur.next = cur2
    
    return head

"""注意：
1.以上方法秩序对链表进行一次遍历，因此时间复杂度为O(N), N表示链表的长度，另外需要几个指针变量来保持结点的地址信息，因此，空间复杂度为O(1)
2.
"""


if __name__ == "__main__":
    head1 = constructList(1)
    head2 = constructList(2)

    print("head1:")
    printList(head1)
    print("\nhead2")
    printList(head2)

    print("\n合并后的链表")
    head = Merge(head1,head2)
    printList(head)
```

```python
"""1.11 如何判断两个单链表（无环）是否相交

题目描述：单链表相交是指两个链表存在完全重合的部分

解题思路
    方法一：hash法
        若两个链表相交，必定有公共结点，结点地址或引用可作为结点的唯一标识，因此判断两个列表中的结点是否有相同的地址或引用来判断判断链表是否相交。
        首先遍历链表head1， 把所有结点放到hashset中， 接着遍历head2， 每遍历一个结点就判断该点的地址在hashSet中是否存在，若存在则相交，否则将链表head2结束，表名不相交
    方法二：首尾相接法，参考1.6
        将两个链表首位项链（例如把链表head1 尾结点链接到head2的头指针中），然后检测这个链表是否存在环，如果存在，则两个链表相交，二换入口结点即为相交的结点。        
    方法三：尾节点法
        若两个结点相交，那么两链表从橡胶垫到结束都是相同，即判断两链表的最后一个结点是否相同即可。
        即先通过一个链表到达尾部，再通过另一个链表到达尾部，则两个链表相交，几下两链表的长度n1, n2， 再遍历一次，长链表结点先出发前进abs(n1-n2)步，之后两链表同时前进，每次一步，相遇的第一个点几位相交第一个点
"""
class LNode:
    def __init__(self,x):
        self.data = x
        self.next = None

# printList(head)

"""
方法功能：判断两个链表是否相交，如果相交则找出交点
输入参数：head1和head2分别为两个链表的头结点
返回值：如果不相交返回None，如果相交返回相交点
"""
def IsIntersect(head1,head2):
    # 1.1判空
    if head1 == None or head1.next == None or head2 == None or head2.next == None or head1 == head2:
        return None

    # 1.2 初始化
    temp1 = head1.next
    temp2 = head2.next
    n1 =0 
    n2 = 0

    # 1.3 遍历，head1，head2 找尾节点并记录head1，head2长度
    while temp1.next != None:
        temp1 = temp1.next
        n1 += 1
    while temp2.next != None:
        temp2 = temp2.next
        n2 += 1
    
    # 1.4 head1 和head2 具有相同的尾节点时
    if temp1 == temp2:
        # 1.4.1长链表先行
        if n1 - n2 > 0:
            while n1-n2 >0:
                head1 = head1.next
                n1 -= 1
        if n2 - n1 > 0:
            while n2-n1 > 0:
                head2 = head2.next
                n2 -= 1

        # 1.4.2 两个链表同时前进，找到相同的结点
        while head1 != head2:
            head1 = head1.next
            head2 = head2.next
        return head1
    # 1.5 head1 和head2 没有相同的尾节点时
    else:
        return None

"""注意：

1.假设两个链表长度分别为n1，n2， 重叠的结点的个数为L（0<L<min(n1,n2)), 则总共对链表进行遍历的次数为:n1+n2+L+n1-L + n2-L = 2(n1 + n2) - L , 因此，算法时间复杂度为O(n1 + n2), 只是用了常数个额外指针变量，因此，空间复杂度为O(1)

"""


if __name__ == "__main__":
    i = 1
    # 链表头结点
    head1 = LNode(None)
    head2 = LNode(None)
    head1.next= None
    head2.next = None

    tmp = None
    cur = head1
    p = None

    # 构造第一个链表
    while i < 25:
        tmp = LNode(None)
        tmp.data = i
        tmp.next = None
        cur.next = tmp
        cur = tmp
        if i==11:
            p = tmp
        i += 1
    cur = head2

    # 构造第二个链表
    i = 1
    while i < 158:
        tmp = LNode(None)
        tmp.data = i
        tmp.next = None
        cur.next = tmp
        cur = tmp
        i += 1
    # 相交于某个点
    cur.next = p
    interNode = IsIntersect(head1,head2)
    if interNode == None:
        print("这两个链表不相交")
    else:
        print("这两个链表相交点：%s" % str(interNode.data))
```
